```python
# Import dependencies
import pandas as pd
from pathlib import Path

# Declare filepath variables for CSVs
census_data_path = "Resources/ACSST1Y2022.S1901-2023-11-29T020523.csv"
cw_data_path = "Resources/CWData.csv"

# Create dataframes from CSVs
census_data = pd.read_csv(census_data_path)
cw_data = pd.read_csv(cw_data_path)
```

# Census Data

```python
# Check work
census_data.head()
```

```python
# Remove columns we don't care about for the overall project
census_data = census_data[census_data.columns.drop(list(census_data.filter(regex='Families')))]
census_data = census_data[census_data.columns.drop(list(census_data.filter(regex='families')))]
census_data = census_data[census_data.columns.drop(list(census_data.filter(regex='Nonfamily')))]
del census_data['Texas!!Households!!Estimate']

# Flip the table
census_data = census_data.transpose()

# Remove new columns we don't care about for the overall project (PERCENT ALLOCATED)
del census_data[13]

# Turn Row 1 values into Column Headers & delete Row 1 containing prev. Column Headers
census_data.rename(columns=census_data.iloc[0], inplace = True)
census_data.drop(census_data.index[0], inplace=True)

# Check work
census_data.head()
```

```python
# Remove 'County, Texas!!Households!!Estimate' from all County names in index
filtered_index = census_data.index.str.split(' ').str[0]
filtered_census_data = census_data.set_index(filtered_index)

# Convert index into Column & reset
filtered_census_data['County'] = filtered_census_data.index
filtered_census_data = filtered_census_data.reset_index()

# Rename columns for ease of later merging & clarity
filtered_census_data = filtered_census_data.rename(columns={'index':'County'})
filtered_census_data = filtered_census_data.rename(columns={'Total':'Total Number of Households'})

# Check work
filtered_census_data.head()
```

```python
# Export clean census data to CSV file
filtered_census_data.to_csv('Resources/clean_census_data.csv')
```

# Child Welfare Data

```python
# Check work
cw_data.head()
```

```python
# Convert Allegation Types into Columns via Pivot
pivoted_df = cw_data.pivot_table(index="County", columns="Type of Allegation", 
                                 values="Allegation Count", aggfunc='first')

# Merge Pivot w/ newly-created Columns with pre-existing full child welfare dataset
combined_cw = pd.merge(pivoted_df, cw_data, how = "right", on=["County", "County"])

# Check work
combined_cw.head()
```

```python
# Drop duplicate rows ==> 1 per County
combined_cw = combined_cw.drop_duplicates(subset = "County", ignore_index = True)

# Turn NaN Cases into 0 Cases
combined_cw = combined_cw.fillna(0) 

# Check work
combined_cw.head()
```

```python
# Export clean child welfare data to CSV file
combined_cw.to_csv('Resources/clean_cw_data.csv')
```

# Merging Datasets

```python
filtered_census_data.head()
```

```python
# Merge Census & Child Welfare data
happy_childhood_data = pd.merge(filtered_census_data, combined_cw, on=["County","County"])

# Drop rows w/ NaN values
happy_childhood_data = happy_childhood_data.dropna()

# Check Work
happy_childhood_data.head()
```

```python
# Export full clean dataset to CSV file for later analysis for overall project
happy_childhood_data.to_csv('Resources/full_clean_dataset.csv')
```
