```python
import pandas as pd
from pathlib import Path 
import matplotlib.pyplot as plt

#create path for CSV data sets
common_allegations_path = r"Resources\full_clean_dataset.csv"


# Store the paths and command to read
common_allegations = pd.read_csv(common_allegations_path)

#display
common_allegations
```

```python
grouped_data = common_allegations.groupby(['County', 'Abandonment', 'Emotional Abuse', 'Labor Trafficking','Medical Neglect','Neglectful Supervision','Physical Abuse','Physical Neglect','Refusal To Accept Parental Responsibility','Sex Trafficking','Sexual Abuse']).size().reset_index(name='Count')
grouped_data.head()
```

```python
# Remove 'Count' column 
if 'Count' in grouped_data.columns:
    grouped_data = grouped_data.drop(columns=['Count'])

# Find the most common allegation type for each county
most_common_allegations = grouped_data.set_index('County').idxmax(axis=1).reset_index(name='Most Common Allegation')

# Display the DataFrame containing the most common allegations for each county
most_common_allegations

```

```python
# Remove 'Count' column if it exists
if 'Count' in grouped_data.columns:
    grouped_data = grouped_data.drop(columns=['Count'])

# Function to find the second most common allegation type for each county
def second_most_common_allegation(row):
    max_allegation = row.idxmax()
    row = row.drop(labels=max_allegation)
    second_most_common = row.idxmax()
    return second_most_common

# Find the second most common allegation type for each county
second_most_common_allegations = grouped_data.set_index('County').apply(second_most_common_allegation, axis=1).reset_index(name='Second Most Common Allegation')

# Display the DataFrame containing the second most common allegations for each county
second_most_common_allegations
```

```python
# Remove 'Count' column 
if 'Count' in grouped_data.columns:
    grouped_data = grouped_data.drop(columns=['Count'])

# Function to find the third most common allegation type for each county
def third_most_common_allegation(row):
    max_allegation = row.idxmax()
    row = row.drop(labels=max_allegation)
    second_most_common = row.idxmax()
    row = row.drop(labels=second_most_common)
    third_most_common = row.idxmax()
    return third_most_common

# Find the third most common allegation type for each county
third_most_common_allegations = grouped_data.set_index('County').apply(third_most_common_allegation, axis=1).reset_index(name='Third Most Common Allegation')

# Display the DataFrame containing the third most common allegations for each county
third_most_common_allegations
```

```python
#merge all three data frames into one 
merged_df = pd.merge(most_common_allegations, second_most_common_allegations, on='County')
top_three_allegations = pd.merge(merged_df, third_most_common_allegations, on='County')

#display the data frame
top_three_allegations
```

```python
# Plotting a bar graph for occurrences per county
grouped_data.reset_index().plot(figsize=(20,8), kind = 'bar', rot=35, x='County', y='Neglectful Supervision', color='skyblue')
plt.title('Neglectful Supervision Incidents per County')
plt.xlabel('Counties')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.show()
```

```python
grouped_data_2 = common_allegations.groupby(['County', 'Abandonment', 'Emotional Abuse', 'Labor Trafficking','Medical Neglect','Neglectful Supervision','Physical Abuse','Physical Neglect','Refusal To Accept Parental Responsibility','Sex Trafficking','Sexual Abuse']).size().reset_index(name='Count')
grouped_data_2.head()
```

```python
# Remove 'County' column if it exists
if 'County' in grouped_data_2.columns:
    grouped_data_2 = grouped_data_2.drop(columns=['County','Count'])


# Aggregate counts for each allegation type across all counties
agg_counts = grouped_data_2.sum()

# Convert the values to numeric to ensure proper sorting
agg_counts = pd.to_numeric(agg_counts, errors='coerce')

# Plotting the cumulative count of incidents per type using a bar graph
plt.figure(figsize=(10, 6))
agg_counts_sorted = agg_counts.sort_values(ascending=False)
agg_counts_sorted.plot(kind='bar', color='skyblue')

plt.title('Incidents per Allegation Type')
plt.xlabel('Allegation Type')
plt.ylabel('# of Incidences')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
