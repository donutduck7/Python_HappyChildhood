```python
import pandas as pd
from pathlib import Path 

#create path for both CSV data sets
census_data_path = r"C:\Users\Tasnia Wahid\Desktop\BERKELEY PROJECT\CLEAN CENSUS DATA.csv"
    
# Store the paths and command to read
census_data = pd.read_csv(census_data_path)
```

```python
census_data.head()
```

```python
print("Columns in census_data:", census_data.columns)
```

```python
import matplotlib.pyplot as plt

# Creating a figure with a specified background color
plt.figure(facecolor='oldlace', figsize=(10, 5))

# Creating a bar plot between median household income and allegation count
#use hatch='/' to make designs inside the bar. 
plt.bar(census_data['Median income (dollars)'], census_data['Allegation Count'], color='skyblue', alpha=0.7, edgecolor='black')

 # Rotating x-axis labels for better visibility
plt.xticks(rotation=90, color='skyblue', fontsize=8) 

#X and Y styling
plt.xlabel('Median Household Income $', color='skyblue', fontdict={'fontfamily': 'arial'})
plt.ylabel('Allegation Count', color='skyblue', fontdict={'fontfamily': 'arial'})

#X and Y formatting
plt.xlabel('Median Household Income $', fontdict={'fontsize': 11, 'fontweight': 'normal', 'fontstyle': 'normal'})
plt.ylabel('Allegation Count', fontdict={'fontsize': 11, 'fontweight': 'normal', 'fontstyle': 'normal'})

#Title formatting
plt.title('Median Household Income vs Allegation Count',color='skyblue', fontdict={'fontfamily': 'arial'})
plt.title('Bar Plot: Median Household Income vs. Allegation Count', fontdict={'fontsize': 11, 'fontweight': 'normal'})
 
# Set the background color of the plot
plt.gca().set_facecolor('oldlace') 

#Output
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
# Creating a figure with a specified background color
plt.figure(facecolor='oldlace', figsize=(10, 5))
# Plotting a bar chart for each county with a single color (pink in this case)
plt.bar(range(len(census_data)), census_data['Allegation Count'], align='center', alpha=0.7, edgecolor='black', color='skyblue')

# Adding county names as labels with proper spacing
plt.xticks(range(len(census_data)), census_data['County'], rotation='vertical', color='skyblue')
# Set the background color of the plot
plt.gca().set_facecolor('oldlace') 
plt.xlabel('County', color='skyblue', size=15)
plt.ylabel('Allegation Count', color='skyblue', size=15)
plt.title('Bar Plot: Allegation Count Per County',color='skyblue', size=15)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating a scatter plot between median household income and allegation count
plt.figure(figsize=(8, 5))
# Creating a figure with a specified background color
plt.figure(facecolor='oldlace', figsize=(10, 5))
plt.scatter(census_data['Median income (dollars)'], census_data['Allegation Count'], color='skyblue', alpha=0.5)
# Set the background color of the plot
plt.gca().set_facecolor('oldlace') 
# Labeling the axes and providing a title
plt.xlabel('Median Household Income ($)', color='skyblue')
plt.ylabel('Allegation Count', color='skyblue')
plt.title('Scatter Plot: Median Household Income vs. Allegation Count', color='skyblue')
plt.xticks(range(len(census_data)), census_data['County'], rotation=90, fontsize=8, color='skyblue')
plt.grid(True,color='pink')
plt.tight_layout()
plt.show()
```
