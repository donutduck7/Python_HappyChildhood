```python
import pandas as pd
from pathlib import Path

census_data_path = "ACSST1Y2022.S1901-2023-11-29T020523.csv"
census_data = pd.read_csv(census_data_path)
census_data.head()
```

```python
## Remove columns we don't care about
census_data = census_data[census_data.columns.drop(list(census_data.filter(regex='Families')))]
census_data = census_data[census_data.columns.drop(list(census_data.filter(regex='families')))]
census_data = census_data[census_data.columns.drop(list(census_data.filter(regex='Nonfamily')))]
del census_data['Texas!!Households!!Estimate']
#flip the table
census_data = census_data.transpose()
#remove new columns we don't care about
del census_data[13]
census_data.head()
```

```python
#census_data.to_csv('mostly_filtered_census_data.csv')
```

```python
filtered_index = census_data.index.str.split(' ').str[0]
filtered_data = census_data.set_index(filtered_index)
filtered_data = filtered_data.rename(index={'Label':'County'})
filtered_data.head()
```

```python
filtered_data.to_csv('clean_census_data.csv')
```
